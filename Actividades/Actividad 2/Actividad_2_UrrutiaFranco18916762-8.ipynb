### 3.1 Crea una señal impulso (delta) y desplazala
Para esto puedes usar alguna librería de python

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el rango de valores de x centrado en 10
x = np.linspace(5, 15, 1000)

# Generar un impulso unitario
impulso = np.zeros_like(x)
impulso[np.abs(x - 10).argmin()] = 1  # Colocar el impulso en x = 10

# Graficar el impulso unitario centrado en 10
plt.figure(figsize=(10, 5))
plt.stem(x, impulso, use_line_collection=True)
plt.title('Impulso Unitario Centrado en 10')
plt.xlabel('x')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()
```

### 3.2 Genere una señal Senoidal y realice la operación de sampleo
Puedes utilizar el método "stem" de matplotlib para plotear el resultado.

```python
import numpy as np
import matplotlib.pyplot as plt

frecuencia_original = 2  # frecuencia
tiempo_inicio = 0  # tiempo inicial
tiempo_final = 1  # tiempo final

# Generar el tiempo para la señal original
tiempo_original = np.linspace(tiempo_inicio, tiempo_final, 1000)

# Generar la señal senoidal original
senal_original = np.sin(2 * np.pi * frecuencia_original * tiempo_original)

# Visualizar la señal original
plt.plot(tiempo_original, senal_original)
plt.title('Señal Senoidal Original de 10 Hz')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()

# Frecuencia de muestreo
frecuencia_muestreo =20 * frecuencia_original

# Calcular el paso de tiempo para el muestreo
paso_tiempo_muestreo = 1 / frecuencia_muestreo

# Generar el nuevo tiempo de muestreo
tiempo_muestreo = np.arange(tiempo_inicio, tiempo_final, paso_tiempo_muestreo)

# Muestrear la señal original en los puntos de tiempo de muestreo
senal_muestreada =  np.sin(2 * np.pi * frecuencia_original * tiempo_muestreo)

# Visualizar la señal muestreada utilizando stem
plt.stem(tiempo_muestreo, senal_muestreada, linefmt='r-', markerfmt='ro', basefmt='r')
plt.title('Señal Senoidal Muestreada al Doble de la Frecuencia')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()
```

### 3.3 Genere una señal Triangular y realice la operación de sampleo

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Definir los parámetros de la señal triangular
A = 1  # amplitud
T = 1  # período

#  señal triangular
t = np.linspace(0, 2*T, 1000)
senal_triangular = signal.sawtooth(2 * np.pi * t / T, width=0.5)

# Visualizar la señal
plt.plot(t, senal_triangular)
plt.title('Señal Triangular')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()

# Frecuencia de muestreo
frecuencia_muestreo = 10  # Hz

# Calcular la cantidad de puntos de muestreo
cantidad_puntos_muestreo = int((2*T) * frecuencia_muestreo)

# Muestrear la señal triangular
senal_muestreada = signal.resample(senal_triangular, cantidad_puntos_muestreo)

# Generar el nuevo tiempo de muestreo
tiempo_muestreo = np.linspace(0, 2*T, cantidad_puntos_muestreo)

# Visualizar la señal muestreada
plt.stem(tiempo_muestreo, senal_muestreada, linefmt='r-', markerfmt='ro', basefmt='r')
plt.title('Señal Triangular Muestreada')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()

```

### 3.4 Realice la operación de sampleo sobre la señal que generaste en el punto 2

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import resample

# Definir el rango de valores de x
x = np.linspace(0, 1, 1000)

# Definir la frecuencia del seno (Hz) y la frecuencia del coseno (Hz)
frecuencia = 5  # Hz
frecuencia = 5  # Hz

# Frecuencias de muestreo a probar multiplicadas por la frecuencia del seno
frecuencias_muestreo = np.array([1.2, 2, 5]) * frecuencia

# Iterar sobre las frecuencias de muestreo
for frecuencia_muestreo in frecuencias_muestreo:
    # Generar los tiempos de muestreo
    tiempos_muestreo = np.arange(0, 1, 1 / frecuencia_muestreo)

    # Samplear la señal en los tiempos de muestreo
    senal_muestreada = np.sin(2*np.pi*frecuencia*tiempos_muestreo) + np.cos(2*np.pi*frecuencia*tiempos_muestreo)

    # Resamplear la señal
    frecuencia_resampleada = 100  # Hz
    senal_resampleada = resample(senal_muestreada, int(len(senal_muestreada) * frecuencia_resampleada / frecuencia_muestreo))

    # Definir el rango de valores de x para la señal resampleada
    x_resampleado = np.linspace(0, 1, len(senal_resampleada))

    # Visualizar la señal original, las muestras y la señal resampleada
    plt.figure(figsize=(10, 7))

    # Señal original
    plt.subplot(3, 1, 1)
    plt.plot(x, np.sin(2*np.pi*frecuencia*x) + np.cos(2*np.pi*frecuencia*x))
    plt.title('Señal Original (Frecuencia del Seno: {} Hz)'.format(frecuencia))
    plt.xlabel('x')
    plt.ylabel('Amplitud')

    # Muestras de la señal
    plt.subplot(3, 1, 2)
    plt.stem(tiempos_muestreo, senal_muestreada, linefmt='r-', markerfmt='ro', basefmt=' ')
    plt.title('Muestras de la Señal (Frecuencia de Muestreo: {} Hz)'.format(frecuencia_muestreo))
    plt.xlabel('x')
    plt.ylabel('Amplitud')

    # Señal resampleada
    plt.subplot(3, 1, 3)
    plt.plot(x_resampleado, senal_resampleada, 'g--')
    plt.title('Señal Resampleada (Frecuencia de Muestreo: {} Hz)'.format(frecuencia_muestreo))
    plt.xlabel('x')
    plt.ylabel('Amplitud')

    plt.tight_layout()
    plt.show()
```

### 3.6 Explica tus hallazgos

El teoriema de Nyquist-Shannon nos da una manera de saber cual es la frecuencia mìnima para muestrear la señal esta es la doble de la frecuencia màxima de la señal en cuestiòn, esto sucede dado que en el dominio de la frecuencia el muestrear al doble es lo mìnimo para que los espectros no se solapen, asì evitando el aliasing.
Luego còmo las frecuencia menor, genera alasing por lo cual la reconstrucciòn de la señal es mala.Para la segunda frecuencia igual a la frecuencia mìnima, la reconstrucciòn es bueno pero al ser minima se produce un aliasing dado que la señal si bien es una señal muy parecida a una continua en realidad no lo es por lo cual el minimo debe ser algo màs alta. Finalmente al usar una frecuencia cinco veces mayor se cumple a cabalidad el criterio por lo cual la reconstrucciòn de esta se da de buena manera.
